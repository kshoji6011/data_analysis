```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
#: read csv
suicides_df = pd.read_csv('samples/master.csv')
```

```python
suicides_df.head()
```

```python
suicides_df.info()
```

```python
suicides_df.columns
```

```python
# Check the transition each year
year_df = suicides_df.groupby('year').sum()
year_df.tail()
```

```python
# 2016 is the midway of counting??
plt.plot(year_df['suicides_no'])
```

```python
sex_df = suicides_df.groupby('sex').sum()
sex_df
```

```python
sns.set_style('whitegrid')
sns.set_palette('Set1')
sex_label = ['female', 'male']
plt.pie(sex_df['suicides_no'], labels=sex_label, autopct="%1.1f%%")
# the suicides number of male is larger than that of female
```

```python
age_df = suicides_df.groupby('age').sum()
age_df
```

```python
sns.set_style('whitegrid')
sns.set_palette('Accent')
age_label = ['15-24years', '25-34years', '35-54years','5-14years', '55-74years', '75+years']
plt.pie(age_df['suicides_no'], labels=age_label, autopct="%1.1f%%")
#: 35-54years is most ratio 
```

```python
generation_df = suicides_df.groupby('generation').sum()
generation_df
```

```python
sns.set_style('whitegrid')
sns.set_palette('Accent_r')
generation_label = [
    'Boomers', 'G.I. Generation', 'Generation X', 'Generation Z', 'Millenials', 'Silent']
plt.pie(generation_df['suicides_no'], labels=generation_label, autopct="%1.1f%%")
```

```python
# check the corrilation
suicides_df.corr()
```

```python
# 
country_sum = suicides_df.groupby('country').sum()
country_sum
```

```python
countries = country_sum.index.format()
```

```python
sns.set()
plt.rcParams["font.size"] = 12
fig = plt.figure(figsize=(20, 20), dpi=100)
ax = fig.add_subplot(1, 1, 1)
plt.bar(countries, country_sum['suicides_no'])
ax.set_xticklabels(countries, rotation=90)
plt.show()
```

<p>Russia, Japan, USA have may suicides number. We should consider the effects of population</p>

```python
plt.rcParams["font.size"] = 12
fig = plt.figure(figsize=(20, 20), dpi=100)
ax = fig.add_subplot(1, 1, 1)
plt.bar(countries, country_sum['suicides/100k pop'])
ax.set_xticklabels(countries, rotation=90)
plt.show()
```

```python
print(country_sum['suicides/100k pop'].idxmax(), country_sum['suicides/100k pop'].max())
```

<p>Russia has the most suicides number</p>
